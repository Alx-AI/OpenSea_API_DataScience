## Exploring OpenSea's Collections

```python
# query and download all collections that exist on OpenSea
def download_collections_info(save_location):
    url = "https://api.opensea.io/api/v1/collections"
    listocollections = []

    for i in range(0, 10):
        querystring = {"offset":str(i*300),
                       "limit":300}
        response = requests.request("GET", url, params=querystring)
        print(i, end=" ")
        if response.status_code != 200:
            print('error')
            print(response.json())
            break

        # Getting collection data
        collections = response.json()
        # storing collections into list
        listocollections.append(collections)
    
    # Flatten everything into one list
    listocollections = [item for sublist in listocollections for item in sublist]
    
    with open(save_location + 'collections'+str(date.today())+r'.pkl', 'wb') as f:
        pickle.dump(listocollections, f)
        
# load most recent saved collections df
def load_collections_info(save_location):
    files = glob.glob(str(save_location)+'collections_????-??-??.pkl')
    with open(max(files, key=os.path.getctime), 'rb') as f:
        return pickle.load(f)
```

```python
save_location = "./data/"
download_collections_info(save_location = save_location)
```
